```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("남양주_총점_충전량_결합.csv", encoding = 'CP949')
df.head(8)
```

```python
'''
건물1 : 1
건물2 : 2
건물3 : 3

도서관 : 1
민원행정기관 : 1

변압기 : 1
전기안전관리자 : 1
전기차 : 3

교통량 : 2
경제인구 : 2
주차면수 : 2
'''
```

```python
weight = [1,2,3,1,1,1,1,3,2,2,2]
```

```python
df2 = df.fillna(0)

df2.head(10)
```

```python
df_y = df2
```

```python
for i in range(len(df_y)) :
    for j in range(1,12,1) :
        df_y.iloc[i,j] = ( df_y.iloc[i,j] - min(df_y.iloc[:, j]) )/ ( max(df_y.iloc[:,j]) - min(df_y.iloc[:, j]) )
df_y.head()
```

```python
df2 = df_y
```

```python
score = []
for i in range(len(df2)) :
    score.append(0)

    for j in range(1,12,1) :
        score[i] += df2.iloc[i, j] * weight[j-1]
        
df2['총점'] = score
```

```python
df2.head()
```

```python
df2['Rank'] = df2['총점'].rank(ascending = False, method = 'min')
```

```python
print( max(df2['경제인구']), min(df2['경제인구']))
```
